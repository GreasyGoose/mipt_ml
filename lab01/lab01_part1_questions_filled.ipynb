$$
x^T \cdot x = x_1^2 + x_2^2 + ... + x_n^2 = \sum_{i=1}^{n} x_i^2;
$$
$$
\frac{d(x^T x)}{dx} = \begin{pmatrix} \frac{d(x^T x)}{dx_1} \\ \frac{d(x^T x)}{dx_2} \\ \vdots \\ \frac{d(x^T x)}{dx_n} \end{pmatrix};

$$
В нашем случае вместо $X$ стоит $X - AS$:
$$
J = || X - AS ||_F^2 = tr((X - AS)(X - AS)^T).

Раскрываем выражение, от которого берется след:

$$

По свойству следа суммы:
$$
J = tr(XX^T - ASX^T - XS^TA^T + ASS^TA^T) = tr(XX^T) - tr(ASX^T) - tr(XS^TA^T) + tr(ASS^TA^T)

3. $$
-\frac{dtr(XS^TA^T)}{dS} = -\frac{dtr(A^TXS^T)}{dS} =  -A^TX;

\frac{dtr(ASS^TA^T)}{dS} = \frac{dtr(A^TASS^T)}{dS} = \frac{dtr(BSS^T)}{dS} = \frac{dtr(S^TBS)}{dS},
$$
где $B=A^TA$ - симметричная матрица,

$$
{B} =  
\begin{pmatrix}

$\alpha$ - угол поворота.

Новые координаты после поворота:
$$
\begin{pmatrix}
{x_{rot}}  \\
{y_{rot}}   \\
\end{pmatrix} =

{x}  \\
{y}   \\
\end{pmatrix}

$$
$$
y_{rot} = x\sin(\alpha) + y\cos(\alpha)
$$
Значение яркости в новых координатах вычисляется путем интерполяции значений в старых координатах.
